```python
import numpy as np               #importing numpy
uber_data["MILES_CAT"]=np. where(uber_data ['MILES*']>100,"Long trip","short trip")     #creating new column as"MILES_CAT" and gives the condition as 
                                                                          # if miles greater than 100 o/p as long trip else short trip
uber_data.head()
```

```python
conditions=[(uber_data ['MILES*']>100),(uber_data ['MILES*']>50) &     
(uber_data ['MILES*']<=100),(uber_data ['MILES*']<=50)]  # adding conditions to newly add column
categories=["long trip","medium trip","short trip"]
uber_data["MILES_CAT"]=np.select(conditions,categories,default ="unknow") 
uber_data.head()
```
