# Group by 

```python
import pandas as pd
import numpy as np

# Sample dataset
data = {
    'Employee': ['Asha', 'Ravi', 'Meena', 'Ravi', 'Asha', 'Meena', 'Ravi', 'Asha'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'],
    'Sales': [20000, 30000, 25000, 40000, 20000, 27000, 35000, 22000],
     'Units': [2, 3, 2, 4, 2, 3, 4, 2],
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar']
}

df = pd.DataFrame(data)
print(df)
```

```python
c=df.groupby('Department')['Sales'].count()   # Count
print("<====================================Count========>")
print(c)
c=df.groupby('Department')['Sales'].sum()
print("<====================================Sum==========>")
print(c)
c=df.groupby('Department')['Sales'].mean()
print("<=====================================Mean==========>")
print(c)
c=df.groupby('Department')['Sales'].median()
print("<====================================Median==========>")
print(c)
c=df.groupby('Department')['Sales'].max()
print("<====================================Max==========>")
print(c)
c=df.groupby('Department')['Sales'].min()
print("<====================================Min==========>")
print(c)
c=df.groupby('Department')['Sales'].nunique()
print("<====================================Unique==========>")
print(c)
```

# 1) Built-in aggregate functions (single-column)

```python
c=df.groupby('Department')['Sales'].agg(['sum','mean','median','count','max','min','nunique'])
c
```

# 2) Multiple aggregations on multiple columns (list / dict style)

## 2.1 List-style (same functions for each column)

```python
c=df.groupby('Department')[['Sales','Units']].agg(['count','sum','mean','max','min'])
c
```

## 2.2 Dict-style (different functions per column)

```python
c=df.groupby('Department').agg({
    'Sales':['sum','mean','max'],
    'Units':['count','max','min']
})
c
```

## 3) Named aggregation (clean column names, recommended) 

```python
c=df.groupby('Department').agg(
    total_sales = ('Sales', 'sum'),
    avg_sales   = ('Sales', 'mean'),
    max_sales   = ('Sales', 'max'),
    units_sold  = ('Units', 'sum'),
    avg_units   = ('Units', 'mean'),
    emp_count   = ('Employee', 'nunique')
).reset_index()
c
```
